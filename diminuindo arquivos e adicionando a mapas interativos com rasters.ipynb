```python
from osgeo import gdal
```

```python
input_raster = r'inserimos o caminho do raster aqui, cujo tamanho será reduzido'

output_raster = r'inserimos o caminho do raster de saída aqui'

mask_shape = r'inserimos o caminho do shapefile aqui, que será usado como máscara para recortar o raster'   

# função para recortar o raster usando a máscara do shapefile:

gdal.Warp(
    output_raster,
    input_raster,
    cutlineDSName=mask_shape,
    cropToCutline=True
)
```

```python
import leafmap
```

```python
# adiciona um mapa com basemap
m = leafmap.Map(center=[-22, -51], zoom=18, height="800px")
m.add_basemap("HYBRID")
m
```

```python
# adiciona raster
m.layers[-1].visible = True
m.add_raster(output_raster, layer_name="Raster Recortado", opacity=0.7,legend_title="Legenda do Raster Recortado", cmap="greens", vmin=0, vmax=255)
```
