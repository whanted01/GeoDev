```python
def calculate_gosavi(green_band_path,nir_band_path,output_gosavi_path):
    #abre os arquivos raster:
    green_band_ds = gdal.Open(green_band_path)

    nir_band_ds = gdal.Open(nir_band_path)

    #transforma os rasters em arrays:
    green_band_data = green_band_ds.ReadAsArray()

    nir_band_data = nir_band_ds.ReadAsArray()

    #calcula gosavi:
    gosavi = (nir_band_data - green_band_data) / (nir_band_data + green_band_data + 0.16)

    #adquire as informações para criação do raster:
    width = green_band_ds.RasterXSize #largura do raster

    eight = green_band_ds.RasterYSize #altura do raster

    projection = green_band_ds.GetProjection() #sistema de coordenadas

    geotransform = green_band_ds.GetGeoTransform() #matriz de transformação

    bands = green_band_ds.RasterCount #número de bandas

    #cria o raster de saída:
    driver = gdal.GetDriverByName('GTiff')

    gosavi_ds = driver.Create(output_gosavi_path, width, eight, bands, gdal.GDT_Float32)

    gosavi_ds.SetGeoTransform(geotransform)

    gosavi_ds.SetProjection(projection)

    #Escreve o resultado da aritmetica de bandas no raster criado:
    gosavi_band = gosavi_ds.GetRasterBand(1)
    gosavi_band.WriteArray(gosavi)

    #calcula a estatistica do raster:
    gosavi_band.ComputeStatistics(True)

    #limpa as variáveis e libera memória:
    green_band_ds = None
    nir_band_ds = None
    gosavi_ds = None
    
#aponta a localização dos arquivos:
    green_band_path = r'C:\Users\caioc\Desktop\pos\mod07\CBERS_4A_WPM_20220902_205_133_L4_BAND2.tif'

    nir_band_path = r'C:\Users\caioc\Desktop\pos\mod07\CBERS_4A_WPM_20220902_205_133_L4_BAND4.tif'

    output_gosavi_path = r'C:\Users\caioc\Desktop\pos\mod07\CBERS_4A_WPM_20220902_205_133_L4_gosavi.tif'

  
```
